```python
import os
import json
import shutil
import glob
from pydub import AudioSegment
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
shutil.copytree(
    "/content/gdrive/MyDrive/Audio Retrieval/data/",
    "/content/gdrive/MyDrive/Audio Retrieval/experimental_data/"
)
```

```python
for dir_num in range(50):
  dir_path = f"/content/gdrive/MyDrive/Audio Retrieval/experimental_data/convo{dir_num}"
  filepath = f"{dir_path}/transcription.json"

  with open(filepath, "r") as file:
    content = file.read()
    convo = json.loads(content)

  with open(f"{dir_path}/voice_preset.json", "r") as file:
    voice_preset = json.loads(file.read())

  for i, dialogue in enumerate(convo):
    actor, speech = list(dialogue.items())[0]
    if actor == "CSR":
      rec_path = f"{dir_path}/clips/{i}.wav"
      audio = AudioSegment.from_file(rec_path, format="wav")

      if voice_preset["CSR"] == "v2/en_speaker_3":
        speed_up_factor = 1.25
      elif voice_preset["CSR"] == "v2/en_speaker_1":
        speed_up_factor = 1.1

      output_audio = audio.speedup(playback_speed=speed_up_factor)
      output_audio.export(rec_path, format="wav")

    elif actor == "Customer":
      rec_path = f"{dir_path}/clips/{i}.wav"
      audio = AudioSegment.from_file(rec_path, format="wav")
      speed_up_factor = 1.1
      output_audio = audio.speedup(playback_speed=speed_up_factor)
      output_audio.export(rec_path, format="wav")
```

```python
for path in glob.glob("/content/gdrive/MyDrive/Audio Retrieval/experimental_data/*/"):
  clip_dir = f"{path}clips/"
  conversation = AudioSegment.empty()
  for i in range(len(os.listdir(clip_dir))):
    audio_segment = AudioSegment.from_wav(f"{clip_dir}{i}.wav")
    conversation += audio_segment
  conversation.export(f"{path}audio.wav", format="wav")
```
